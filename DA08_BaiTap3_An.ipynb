## Đề Bài

1. Vẽ biểu đồ dữ liệu chứng khoán với nhiều mã khác nhau.

## Script

### Using

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Bài 1

```python
# Read excel files of stock and save to dataframes
df_acb = pd.read_csv('data-stock/excel_acb.csv')
df_dxg = pd.read_csv('data-stock/excel_dxg.csv')
df_flc = pd.read_csv('data-stock/excel_flc.csv')
df_fpt = pd.read_csv('data-stock/excel_fpt.csv')
df_vjc = pd.read_csv('data-stock/excel_vjc.csv')
df_vnm = pd.read_csv('data-stock/excel_vnm.csv')
```

```python
# Combine all dataframes
df = pd.concat([df_acb, df_dxg, df_flc, df_fpt, df_vjc, df_vnm])

# Rename columns
df = df.rename(columns={'<DTYYYYMMDD>': 'Date', '<Close>': 'Price', '<Ticker>': 'Name'})

# Convert the column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
```

```python
# Pivot the dataframe
df_pivot = df.pivot(index='Date', columns='Name', values='Price')

# Plot the pivot table
df_pivot.plot(figsize=(20, 10))

# Set the x and y axis labels and the title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices')

# Display the plot
plt.show()
```

### Bài 2

```python
# Set constant
x_min = -5
x_max = 5
```

```python
# Float try compare to int
def float_try_int(num=0):
    temp = int(num)
    if temp == float(num):
        return temp
    return num
```

```python
# Multiplication of the two parameters by numpy
def multi_mod(a, b):
    return np.multiply(a, b)
```

```python
# Get random number
def get_random_num():
    return round(random.uniform(-100, 100), 1)
```

```python
# Get random values of a, b and c
a = get_random_num()
b = get_random_num()
c = get_random_num()

# Draw horizontal and vertical lines
plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')

# Set the x and y axis labels and the title
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Đồ thị của phương trình y = {float_try_int(a)}x² {b > 0 and "+" or "-"} {float_try_int(abs(b))}x {c > 0 and "+" or "-"} {float_try_int(abs(c))}')

# Plot the graph
x = np.linspace(x_min, x_max, 100)
y = a * x**2 + b * x + c
plt.plot(x, y, linewidth=2)

# Set step
plt.xticks(np.arange(-5, 6, step=1))

# Add grid
plt.grid(linestyle='--', alpha=0.5)

# display the plot
plt.show()
```
