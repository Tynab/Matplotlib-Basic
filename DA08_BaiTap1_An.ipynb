# BÀI TẬP MATPLOTLIB CƠ BẢN

## Đề Bài

Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a, b).

## Script

### Using

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Extension Methods

```python
# Ensure that the input is always an number
def get_input_num(cap='Nhập số: '):
    while True:
        try:
            return float(input(cap))
        except:
            print('Dữ liệu nhập vào phải là số, xin vui lòng thử lại...')
```

### Contract Methods

```python
# Main
def main():
    # import the values of a and b
    a = get_input_num("Nhập a = ")
    b = get_input_num("Nhập b = ")
    # draw horizontal and vertical lines
    plt.axhline(y=0, color='gray')
    plt.axvline(x=0, color='gray')
    # set the x and y axis labels and the title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Đồ thị của phương trình y = ax + b')
    # set constant
    lim = 10
    freq = 100
    # compute the range
    if a == 0:
        bTemp = abs(b)
        pMin, pMax = -bTemp * 2, bTemp * 2
    else:
        xMin, xMax = (-b - lim) / a, (-b + lim) / a
        yMin, yMax = -lim * a + b, lim * a + b
        pMin, pMax = (xMax - xMin > yMax - yMin) and (xMin, xMax) or (yMin, yMax)
    # set the limit
    plt.xlim(pMin, pMax)
    plt.ylim(pMin, pMax)
    # plot the graph
    x = np.linspace(pMin, pMax, freq)
    plt.plot(x, np.add(np.multiply(a, x), b), linewidth=2)
    # add grid
    plt.grid(linestyle='--', alpha=0.5)
    # set the aspect ratio
    plt.gca().set_aspect('equal')
    # display the plot
    plt.show()
```

```python
# Run
main()
```
