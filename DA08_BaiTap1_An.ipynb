# BÀI TẬP MATPLOTLIB CƠ BẢN

## Đề Bài

Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a, b).

## Script

### Using

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Extension Methods

```python
# Ensure that the input is always an number
def get_input_num(cap='Nhập số: '):
    while True:
        try:
            return float(input(cap))
        except:
            print('Dữ liệu nhập vào phải là số, xin vui lòng thử lại...')
```

```python
# Float try compare to int
def float_try_int(num=0):
    temp = int(num)
    if temp == float(num):
        return temp
    return num
```

### Contract Methods

```python
# Main
def main():
    # import the values of a and b
    a = get_input_num("Nhập a = ")
    b = get_input_num("Nhập b = ")
    # draw horizontal and vertical lines
    plt.axhline(y=0, color='gray')
    plt.axvline(x=0, color='gray')
    # set the x and y axis labels and the title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Đồ thị của phương trình y = {float_try_int(a)}x {b > 0 and "+" or "-"} {float_try_int(abs(b))}')
    # set the limit
    plt.xlim(-5, 5)
    # plot the graph
    x = np.linspace(-6, 6, 100)
    y = a * x + b
    plt.plot(x, y, linewidth=2)
    # add grid
    plt.grid(linestyle='--', alpha=0.5)
    # display the plot
    plt.show()
```

```python
# Run
main()
```
