## Script

### Using

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Contract Fields

```python
# Read excel files of stock and save to dataframes
df_acb = pd.read_csv('data-stock/excel_acb.csv')
df_dxg = pd.read_csv('data-stock/excel_dxg.csv')
df_flc = pd.read_csv('data-stock/excel_flc.csv')
df_fpt = pd.read_csv('data-stock/excel_fpt.csv')
df_vjc = pd.read_csv('data-stock/excel_vjc.csv')
df_vnm = pd.read_csv('data-stock/excel_vnm.csv')
```

```python
dfs = [df_acb, df_dxg, df_flc, df_fpt, df_vjc, df_vnm]
```

### Processing

```python
# Convert column to datetime for each element
for df in dfs:
    df['<DTYYYYMMDD>'] = pd.to_datetime(df['<DTYYYYMMDD>'], format='%Y%m%d')
```

```python
# Get limit
max_close = max(df['<Close>'].max() for df in dfs)
min_date = min(df['<DTYYYYMMDD>'].min() for df in dfs)
max_date = max(df['<DTYYYYMMDD>'].max() for df in dfs)
```

### Drawing

```python
# Set size
plt.subplots(figsize=(20, 10))

# Draw each element
for i, df in enumerate(dfs):
    # Focus path
    plt.subplot(2, 3, i + 1)
    # plot the graph
    plt.plot(df['<DTYYYYMMDD>'], df['<Close>'])
    # set the x and y axis labels and the title
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(df['<Ticker>'][0])
    # set the limit
    plt.xlim(min_date, max_date)
    plt.ylim(0, max_close)

# Fit all element
plt.tight_layout()

# Display the plot
plt.show()
```
